```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

imgL = cv2.imread('data/l.png',0)
imgR = cv2.imread('data/r.png',0)

stereo = cv2.StereoSGBM_create(numDisparities=96, blockSize=7)
disparity = stereo.compute(imgL,imgR)
print(disparity)
plt.imshow(disparity,'gray')
plt.show()
```

```python
b = 0.54
f = 718.856

distance = (b*f)/(disparity)
print(distance)
```
